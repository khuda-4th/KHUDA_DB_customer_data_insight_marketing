RFM 고객 세그먼트 분류를 해보자!

```python
#1. 세 항목에 동일하게 가중치를 줘 보자.
weight = {}
weight['recency'] = 1 / 3
weight['frequency'] = 1 / 3
weight['monetary'] = 1 / 3
```

```python
#등급을 숫자형 데이터로 바꿔주기
df['rfm_score'] = (
    weight['recency'] * df['recency_grade'].astype('int')
    + weight['frequency'] * df['frequency_grade'].astype('int')
    + weight['monetary'] * df['monetary_grade'].astype('int')
)
```

```python
#RFM등급을 5개로 세분화 시키자
def rfm_segment_bins(x):
    if x < 9 / 5:
        return 1
    elif x <= 13 / 5:
        return 2
    elif x <= 17/5:
        return 3
    elif x <= 21/5:
        return 4
    else:
        return 5
```

```python
df['rfm_segment'] = df['rfm_score'].apply(rfm_segment_bins)
```

```python
rfm_counts =  df['rfm_segment'].value_counts()
rfm_counts
```

```python
sns.barplot(x = rfm_counts.index, y = rfm_counts.values)
```

```python
#3등급 고객 수가 가장 많고 1,5등급으로 갈 수록 적어진다
```

rfm 고객 세그먼트 별 매출 기여도 확인하기


```python
groupby_rfm_segment = df.groupby('rfm_segment').mean(numeric_only=True).reset_index()
groupby_rfm_segment['amount_total'].plot(
    kind='pie',
    autopct='%.1f%%',
    labels=[f'{x}등급' for x in [1,2,3,4,5]],
    title='RFM 고객 세그먼트별 매출 기여도',
    ylabel='',
)
```

```python
#3,4,5등급이 전체 약 95%로 대부분이다
#recency는 등급별 편차가 거의 없었으므로 recency의 가중치를 줄이자
# 가중치 재설정
weight['recency'] = 0.2
weight['frequency'] = 0.4
weight['monetary'] = 0.4

# 가중합 계산
df['rfm_score'] = (
    weight['recency'] * df['recency_grade'].astype('int')
    + weight['frequency'] * df['frequency_grade'].astype('int')
    + weight['monetary'] * df['monetary_grade'].astype('int')
)

# RFM 고객 지표 구간화
df['rfm_segment'] = df['rfm_score'].apply(rfm_segment_bins)
```

```python
rfm_counts = df['rfm_segment'].value_counts()
rfm_counts
```

```python
sns.barplot(x = rfm_counts.index, y = rfm_counts.values)
```

```python
#아까랑은 사뭇 다른 결과이다. 4등급 고객 수가 가장 높다. 
#세그먼트별 매출 기여도 시각화
groupby_rfm_segment = df.groupby('rfm_segment').sum(numeric_only=True).reset_index()
groupby_rfm_segment['amount_total'].plot(
    kind='pie',
    autopct='%.1f%%',
    labels=[f'{x}등급' for x in [1,2,3,4,5]],
    title='RFM 고객 세그먼트별 매출 기여도',
    ylabel='',
)
```
